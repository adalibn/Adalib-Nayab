# **‚≠ê Classification Model to Identify Multiple Disease**

```python
# import library
import pandas as pd
```

```python
# import data
disease = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/MultipleDiseasePrediction.csv')
```

```python
# view data
disease.head()
```

```python
# info of data
disease.info()
```

```python
# summary statistics
disease.describe()
```

```python
# check for missing value
disease.isnull().sum()
```

```python
# check for categories
disease.nunique()

```

```python
# correlation
disease.corr()
import matplotlib.pyplot as plt
plt.figure(figsize = (15,12))
import seaborn as sns
sns.heatmap(disease.corr(),annot = True)
```

```python
# visualize pairplot
import seaborn as sns
sns.pairplot(disease)
```

```python
# column names
disease.columns
```

```python
# define y
y = disease['prognosis']
```

```python
y.shape
```

```python
# define X
X = disease.drop(columns = 'prognosis',axis = 1)
```

```python
# split data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=2529)
```

```python
# verify shape
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
# select model
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(criterion='entropy')
```

```python
# train model
model.fit(X_train,y_train)
```

```python
# predict with model
y_pred = model.predict(X_test)
```

```python
# model evaluation
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
# model accuracy
accuracy_score(y_pred,y_test)
```

```python
# model confusion matrix
confusion_matrix(y_pred,y_test)
```

```python
# model classification report
print(classification_report(y_pred,y_test))
```
